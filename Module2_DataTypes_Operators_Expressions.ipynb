#### --- Python Data Types: Integers, Strings, Booleans ---

```python
# Integer
num = 10
print(f"Integer: {num}, {type(num)}")

# Float
float_num=3.14
print("Float:", float_num, type(float_num))
# String
name = ""
print(f"String: {name}, {type(name)}")

sentence='''This is a Python Programming.
It supports multiple lines.'''
print(f"String (multiline): {sentence}, {type(sentence)}")

# Boolean
is_active = True
print(f"Boolean: {is_active}, {type(is_active)}")
```

#### --- Arithmetic Operators ---
#### +, -, *, /, //, %, **

```python
a = 15
b = 4

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Exponentiation:", a ** b)
```

#### --- Comparison (Relational) Operators ---
#### ==, !=, >, <, >=, <=

```python
x = 10
y = 20

print("x == y:", x == y)
print("x != y:", x != y)
print("x > y:", x > y)
print("x < y:", x < y)
print("x >= y:", x >= y)
print("x <= y:", x <= y)
```

#### --- Assignment Operators ---
#### =, +=, -=, *=, /=, %=, **=, //=

```python
num = 25
print("Initial value:", num)

num += 3
print("After += 3:", num)

num -= 2
print("After -= 2:", num)

num *= 4
print("After *= 4:", num)

num /= 2
print("After /= 2:", num)

num %= 3
print("After %= 3:", num)

num **= 2
print("After **= 2:", num)

num //= 2
print("After //= 2:", num)
```

#### --- Logical Operators ---
#### and, or, not

```python
a = True
b = False

print("a and b:", a and b)
print("a or b:", a or b)
print("not a:", not a)
```

#### --- Bitwise Operators ---
#### &, |, ^, ~, <<, >>

```python
x = 10   # binary: 1010
y = 4    # binary: 0100

print("x & y:", x & y)   # AND
```

```python
print("x | y:", x | y)   # OR
```

#### Returns 1 if the bits are different, else 0.
####   1010
#### ^ 0100
####   ----
####   1110

```python
print("x ^ y:", x ^ y)   # XOR
```

### NOT
### Flips all bits.
### In binary, ~x = -(x + 1)

```python

print("~x:", ~x)         # NOT
```

### Definition:
### Shifts bits to the left by the specified number of positions.
### Each left shift multiplies the number by 2ⁿ (where n = number of shifts).
###   1010  → 101000
### O/P: 40 (10 * 2² = 40)

```python


print("x << 2:", x << 2) # Left Shift
```

### Definition:
### Shifts bits to the right by the specified number of positions.
### Each right shift divides the number by 2ⁿ (floor division).
### Output: 2 (10 // 2² = 2)

```python

print("x >> 2:", x >> 2) # Right Shift
```

### --- Expressions ---
### Expressions combine values and operators to produce a result

```python
# --- Expressions ---
# Expressions combine values and operators to produce a result

x = 5
y = 10
result = (x + y) * 2 - (y / x)

print("Expression result:", result)
```
