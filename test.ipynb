```python
## Variable Length Arguments
## Positional & Keywords arguments

def print_numbers(*args):
    for numbers in args:
        print(numbers)
```

```python
print_numbers(1,2,3,4,'balaji')
```

```python
## Keywords Arguments
def print_details(**kwargs):
    for key, value in kwargs.items():
        print(f'Key: {key}, Value: {value}')
print_details(name='Balaji', age='28', Country='India')
```

```python
print_details(name='Balaji', age='28', Country='India')
```

```python
def print_details(*args, **kwargs):
    for val in args:
        print(f"Positional arguments: {val}")

    for key, value in kwargs.items():
        print(f'Key: {key}, Value: {value}')
```

```python
print_details(1,2,3,4,'balaji',name='Balaji', age='27', Country='India')
```
