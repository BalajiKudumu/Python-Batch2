### Types of Methods in a Class

1. <b>Instance Method</b> - Works with object data (uses self)

2. <b>Class Method</b> - Works with class data (uses cls)

3. <b>Static Method</b> - General method, doesn’t need class or object reference

```python
class Employee:
    company = "TechCorp"   # class variable

    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def show(self):   # Instance method
        print(f"{self.name} works in {Employee.company}")

    @classmethod
    def change_company(cls, new_name):
        cls.company = new_name

    @staticmethod
    def greet():
        print("Welcome to the company!")

e1 = Employee("Raj", 150000)
e1.show()            # Raj works in TechCorp
Employee.change_company("InnoTech")
e1.show()            # Raj works in InnoTech
Employee.greet()     # Welcome to the company!
```

```python
class BankAccount:
    bank_name = "SBI"

    def __init__(self, name, balance):
        self.name = name
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"{amount} deposited. New balance: {self.balance}")

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            print(f"{amount} withdrawn. New balance: {self.balance}")
        else:
            print("Insufficient balance!")

    @classmethod
    def change_bank(cls, new_bank):
        cls.bank_name = new_bank

    @staticmethod
    def info():
        print("Welcome to our banking system!")

# Using class
BankAccount.info()

# Creating objects
acc1 = BankAccount("Balaji", 7000)
acc1.deposit(500)
acc1.withdraw(300)

# Changing bank name using class method
BankAccount.change_bank("HDFC")
print("New Bank:", BankAccount.bank_name)
```

### INHERITANCE

#### Inheritance means one class (child/subclass) can use the properties and methods of another class (parent/base class).

### Single inheritance class B(A)

```python

# Parent class 
class Animal:
    def speak(self):
        print("Animal makes a sound")

# Child class
class Dog(Animal):
    def bark(self):
        print("Dog barks")

# Create object of Dog
d = Dog()
d.speak()   # Inherited from Animal
d.bark()    # Defined in Dog
```

### Multilevel Inheritance
#### class C(B) - > class B(A) (Parent -> Child -> Grandchild)

```python
class GrandFather:
    def land(self):
        print("GrandFather has land.")

class Father(GrandFather):
    def car(self):
        print("Father has a car.")

class Son(Father):
    def bike(self):
        print("Son has a bike.")

s = Son()
s.land()
s.car()
s.bike()
```

### Multiple Inheritance
#### Child inherits from multiple parents class C(A, B):

```python
class Father():
    def land(self):
        print('Father has land')

class Mother():
    def money(self):
        print('Mother has money')

class Child(Father, Mother):
    def bike(self):
        print('Child has bike')

c=Child()
c.land()
c.money()
c.bike()
```

### Hierarchical Inheritance
#### One parent -> Many children class B(A), class C(A)

```python
class Father():
    def land(self):
        print('Father has land')

class child1(Father):
    def money(self):
        print('child1 has money')

class Child2(Father):
    def bike(self):
        print('Child2 has bike')

c1=child1()
c1.land()
c1.money()

c2=Child2()
c2.land()
c2.bike
```

### METHOD OVERRIDING

#### Method Overriding happens when a child class defines a method with the same name as the parent class method.
#### The child’s version replaces (overrides) the parent’s method.

```python
class Animal:
    def sound(self):
        print("Animals make different sounds")

class Dog(Animal):
    def sound(self):
        print("Dog barks")  # overriding parent method

d = Dog()
d.sound()   # Output: Dog barks
```

#### Using super() to call the parent method

#### Sometimes, we want to call both the parent and child methods.

```python
class Animal:
    def sound(self):
        print("Animals make sound")

class Dog(Animal):
    def sound(self):
        super().sound()    # call parent method
        print("Dog barks")

d = Dog()
d.sound()
```

### DATA HIDING 
#### Data Hiding means restricting access to the internal details of an object.
#### In Python, this is done by using private variables (with __ before the name).

#### This prevents direct modification of important data from outside the class.

```python
class BankAccount():
    def __init__(self, owner, balance=0):
        self.owner=owner
        self.__balance=balance


    def getBalance(self):
        return self.__balance
    

# Create a object
account=BankAccount('Balaji Kudumu',173000)
print(account.getBalance())
# print(account.__balance)
```

```python
class Account:
    def __init__(self, name, balance):
        self.name = name
        self.__balance = balance   # private variable

    def deposit(self, amount):
        self.__balance += amount

    def show_balance(self):
        print(f"Account balance: {self.__balance}")

acc = Account("Balaji", 1000)

acc.deposit(500)
acc.show_balance()       # Output: Account balance: 1500

# Trying to access private variable directly
print(acc.__balance)     # Error: AttributeError
```

### Accessing Private Data Safely
#### We can read or modify private variables using getter/setter methods

```python
class Account:
    def __init__(self, name, balance):
        self.__name = name
        self.__balance = balance

    # getter
    def get_balance(self):
        return self.__balance

    # setter
    def set_balance(self, new_balance):
        if new_balance >= 0:
            self.__balance = new_balance
        else:
            print("Invalid balance!")

acc = Account("Raj", 2000)
print(acc.get_balance())   # 2000
acc.set_balance(3000)
print(acc.get_balance())   # 3000
acc.set_balance(-500)      # Invalid balance!

print(f"Updated Balance: {acc.get_balance()}")
```

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.__salary = salary  # private variable

    def show(self):
        print(f"Employee: {self.name}, Salary: {self.__salary}")

class Developer(Employee):  # Inheritance
    def __init__(self, name, salary, project):
        super().__init__(name, salary)  # call parent constructor/
        self.project = project

    def show(self):  # Method overriding
        super().show()
        print(f"Project: {self.project}")

# Create object
dev = Developer("John", 180000, "AI Engineer")
dev.show()
```

### User-Defined Exceptions

#### You can define your own custom exceptions by creating a class that inherits from Exception.

```python
class InsufficientBalanceError(Exception):
    """Custom Exception for low balance"""
    pass

class BankAccount:
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance

    def withdraw(self, amount):
        if amount > self.balance:
            raise InsufficientBalanceError("Insufficient balance to withdraw.")
        else:
            self.balance -= amount
            print(f"Withdrawal successful! Remaining balance: {self.balance}")

# Using our class
try:
    acc = BankAccount("Balaji", 1000)
    acc.withdraw(1500)
except InsufficientBalanceError as e:
    print("Error:", e)
finally:
    print("Transaction completed.")
```
