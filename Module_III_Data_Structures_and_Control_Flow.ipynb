#### --- Tuples ---
#### Tuples are immutable sequences

```python
my_tuple = (10, 20, 30, 40)
print("Tuple:", my_tuple)
print("First Element:", my_tuple[0])
print("Slice:", my_tuple[1:3])
```

#### --- Sets ---
#### Sets store unique elements and are unordered

```python

set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

# Uninon
print("Union:", set1 | set2)
```

```python
# Intersection
print("Intersection:", set1 & set2)

# Difference
print("Difference:", set1 - set2)

#Symmetric Difference
print("Symmetric Difference:", set1 ^ set2)
```

```python
# add method
set1.add(10)
print(set1)
```

```python
set1.remove(1)
print("Modified Set1:", set1)
```

#### --- For Loops ---

```python
# Using range()
for i in range(1, 6):
    print("Range Value:", i)
```

```python
# Loop through list
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print("Fruit:", fruit)
```

```python
# Loop through dictionary
student = {"name": "Alice", "marks": 90, "course": "Python"}
print(student.items())
```

```python
for key, value in student.items():
    print(f"{key}: {value}")
```

#### --- List Comprehensions ---

```python
squares = [x**2 for x in range(1, 11)]
print("Squares:", squares)
```

```python
even_numbers = [x for x in range(1, 11) if x % 2 == 0]
print("Even Numbers:", even_numbers)
```

```python
fruits_upper = [fruit.upper() for fruit in ["apple", "banana", "cherry"]]
print("Uppercase Fruits:", fruits_upper)
```

#### --- While Loop ---

```python

count = 0
while count < 5:
    print("Count:", count)
    count += 1
else:
    print("Loop finished successfully")
```

```python
# Break statement (exits the loop)
for i in range(1, 10):
    if i==3:
        break
    print(i)
```

```python
# Continue (Skips the current iteration and moves to the next)
for i in range(1, 10):
    if i==3:
        continue # Skips 
    print(i)
```

```python
# Pass (Does nothing, acts as a placeholder) for future development
for i in range(1, 10):
    if i==3:
        pass # Skips 
    print(i)
```

#### --- Nested Conditional and Loop Blocks ---

```python
numbers = [1, 2, 3, 4, 5]
for num in numbers:
    if num % 2 == 0:
        print(num, "is Even")
    else:
        print(num, "is Odd")
else:
    print("Completed checking all numbers.")
```

#### --- Example ---

```python
students = [
    {"name": "Alice", "marks": 85},
    {"name": "Bob", "marks": 72},
    {"name": "Charlie", "marks": 90},
]

grades = {}

for student in students:
    name = student["name"]
    marks = student["marks"]

    if marks >= 85:
        grade = "A"
    elif marks >= 70:
        grade = "B"
    else:
        grade = "C"

    grades[name] = grade

print("Final Grades:")
for name, grade in grades.items():
    print(f"{name}: {grade}")
```
